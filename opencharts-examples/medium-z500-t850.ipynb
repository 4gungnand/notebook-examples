## Retrieve Data
This product takes in input 2 parameters :

* [Geopotential height](https://apps.ecmwf.int/codes/grib/param-db/?id=156)
* [Temperature](https://apps.ecmwf.int/codes/grib/param-db/?id=130)

In this example, we will use: 
- **ecmwf.opendata** Client to download the data
- **ecmwf.data** library to read and process the data (calculate the speed and shear)
- **ecmwf.geomaps** to plot the result

```python
import ecmwf.data as ecdata
from ecmwf.geomaps import GeoMap
from ecmwf.opendata import Client
```

```python
client = Client("https://ai4edataeuwest.blob.core.windows.net/ecmwf", beta=True)
```

```python
parameters = ['gh', 't']
filename = 'medium-z500-t850.grib'
filename
```

```python
client.retrieve(
    date=-2,
    time=0,
    step=12,
    stream="oper",
    type="fc",
    levtype="pl",
    levelist=[500,850],
    param=parameters,
    target=filename
)
```

## Reading and processing the data
Now we can use **ecmwf.data** to read the file.

```python
data = ecdata.read(filename)
```

The **describe()** function will give us the overview of the dataset.  

```python
data.describe()
```

And an overview of one parameter

```python
data.describe('gh')
```

We can use **ls()** function to list all the fields in the file we downloaded.

```python
data.ls()
```

The grib file contains all the parameters and levels, and we will use the **select()** function to filter what we need.

```python
t850 = data.select(shortName= 't',level=850)
t850.describe()
```

<div class="alert alert-block alert-danger">
<b>Don't forget:</b> /100 will change once we sort out scalling for GH
</div>

```python
gh500 = data.select(shortName= 'gh',level=500)
gh500.describe()
```

Geopotential height has units gpm (geopotential meters), but on the ECMWF Open charts it is plotted in geopotential decameters. To reproduce the plot we need to divide by 100.

```python
gh500 /= 100
```

And finally, we can plot the data on the map. 

```python
fig = GeoMap(area_name='europe', preset='background-foreground')

fig.contour_lines(t850, preset="sh_all_fM80t56i4")
fig.contour_lines(gh500, preset="ct_blk_i4_t2")

fig.title(["Geopotential 500 hPa and temperature at 850 hPa", 
           "START TIME: <grib_info key='base-date' format='%a %d %B %Y %H' where='shortName=t'/> ",
           "VALID TIME: <grib_info key='valid-date' format='%a %d %B %Y %H' where='shortName=t'/>, STEP: <grib_info key='step' where='shortName=t' />"])
fig.show()
```
