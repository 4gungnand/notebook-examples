**MARS retrieval to get ERA 5 data**

To get data https://software.ecmwf.int/wiki/display/CKB/How+to+download+ERA5+data+via+the+ECMWF+Web+API


```python
from ecmwfapi import ECMWFDataServer
 
server = ECMWFDataServer()
server.retrieve({
    "class": "ea",
    "dataset": "era5",
    "expver": "1",
    "stream": "oper",
    "type": "an",
    "levtype": "sfc",
    "param": "167.128",
    "date": "2016-01-01/to/2016-01-01",
    "time": "00:00:00",
    "step": "0",
    "grid": "0.25/0.25",
    "format": "netcdf",
    "target": "test.nc"
 })
```

```python
import Magics.macro as ma
```

```python
!ncdump test.nc
```

```python
data = ma.mnetcdf(
  netcdf_type = "geomatrix",
  netcdf_filename = "test.nc",
  netcdf_value_variable = "t2m",
  netcdf_latitude_variable = "latitude",
  netcdf_longitude_variable = "longitude"
)
```

```python
contour = ma.mcont(contour_line_colour = "red")
ma.plot(ma.mcoast(),data,contour)
```
