## Retrieve Data
This product takes in input 1 parameter:

* [Geopotential height](https://apps.ecmwf.int/codes/grib/param-db/?id=156)

In this example, we will use: 
- **ecmwf.opendata** Client to download the data
- **ecmwf.data** library to read and process the data (calculate the speed and probabilities)
- **ecmwf.geomaps** to plot the result

```python
import ecmwf.data as ecdata
from ecmwf.geomaps import GeoMap
from ecmwf.opendata import Client
```

```python
client = Client("ecmwf", beta=True)
```

For the geopotential height at 500 hPa, ensemble mean and spread are available for download.  

Ensemble mean and ensemble spread have different **type** in the request.  
One data request can have only one type, so we need to have two requests to download this datasets.

```python
parameters = ['gh']
em_filename = 'medium-t500-mean-spread_em.grib'
es_filename = 'medium-t500-mean-spread_es.grib'
em_filename
```

```python
client.retrieve(
    date=-2,
    time=0,
    step=144,
    stream="enfo",
    type="em",
    levtype="pl",
    levelist=[500],
    param=parameters,
    target=em_filename
)
```

```python
client.retrieve(
    date=-2,
    time=0,
    step=144,
    stream="enfo",
    type="es",
    levtype="pl",
    levelist=[500],
    param=parameters,
    target=es_filename
)
```

## Reading and processing the data
Now we can use **ecmwf.data** to read the file.

```python
gh_em = ecdata.read(em_filename)
```

```python
gh_es = ecdata.read(es_filename)
```

The **describe()** function will give us the overview of the dataset.  

```python
gh_em.describe('gh')
```

```python
gh_es.describe()
```

We can use **ls()** function to list all the fields in the file we downloaded.

```python
gh_es.ls()
```

Geopotential height has units gpm (geopotential meters), but on the ECMWF Open charts it is plotted in geopotential decameters. To reproduce the plot we need to divide by 10.

```python
gh_em /= 10
```

```python
gh_es = gh_es/10
```

## Plotting the data
And finally, we can plot the data on the map. 

```python
fig = GeoMap(area_name='europe', preset='background-foreground')

fig.contour_lines(gh_es, preset="sh_blu_f02t50")
fig.contour_lines(gh_em, preset="ct_red_i5_t2")

fig.title(["Ensemble mean and spread for 300 hPa geopotential", 
           "START TIME: <grib_info key='base-date' format='%a %d %B %Y %H' />", 
           "VALID TIME: <grib_info key='valid-date' format='%a %d %B %Y %H' />, STEP: <grib_info key='step' />"])
fig.legend()

fig.show()
```
