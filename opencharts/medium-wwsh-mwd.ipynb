## Retrieve Data
This product takes in input 3 parameters :

* [	Significant height of wind waves](https://apps.ecmwf.int/codes/grib/param-db/?id=140234)
* [Mean direction of wind waves](https://apps.ecmwf.int/codes/grib/param-db/?id=140235)

In this example, we will use: 
- [**ecmwf-api-client**](https://pypi.org/project/ecmwf-api-client/) to download the data
- [**ecmwf.data**](https://github.com/ecmwf/ecmwf-data) library to read and process the data 
- [**magpye**](https://magpye.readthedocs.io) to plot the result

```python
import ecmwf.data as ecdata
from magpye import GeoMap
from ecmwfapi import *
import requests
```

```python
parameters = ["140234/140235"]
filename = 'medium-wwsh-mwd.grib'
filename
```

```python
server = ECMWFService("mars")
server.execute(
    {
    "class": "od",
    "date": "-1",
    "expver": "1",
    "levtype": "sfc",
    "param": parameters,
    "step": "12",
    "stream": "wave",
    "time": "00",
    "type": "fc",
    "grid": "0.4/0.4"
    },
    filename)
```

Alternatevly we have prepared a small sample dataset to explore

```python
url = f"https://get.ecmwf.int/repository/opencharts-sample-data/{filename}"
r = requests.get(url)
with open(filename, 'wb') as file:
    file.write(r.content)
```

## Reading and processing the data
Now we can use **ecmwf.data** to read the files.

```python
data = ecdata.read(filename)
```

The **describe()** function will give us the overview of the dataset.  

```python
data.describe()
```

And an overview of one parameter, where we can see more information, such as units or type of level.

```python
data.describe('mdww')
```

We can use **ls()** function to list all the fields in the file we downloaded.

```python
data.ls()
```

The grib file contains all the parameters, and we will use the **select()** function to filter what we need.  
Feel free to use **describe()** to inspect the other filtered fields.

```python
shww = data.select(shortName= "shww")
mdww = data.select(shortName= "mdww")
```

And finally, we can plot the data on the map. 

```python
fig = GeoMap(area_name='north_atlantic')

fig.coastlines(land_colour="cream",resolution="medium")

fig.contour_shaded(shww, style="waves_rainbow_20")
fig.arrows(speed=shww, direction=mdww, wind_style="arrows", density=2, colour='black')

fig.coastlines(resolution="medium")
fig.gridlines()

fig.title(["Windsea: Significant wave height and mean direction",
           "START TIME: <grib_info key='base-date' format='%a %d %B %Y %H' where='shortName=shww'/> ",
           "VALID TIME: <grib_info key='valid-date' format='%a %d %B %Y %H' where='shortName=shww' />, STEP: <grib_info key='step' where='shortName=shww'/>"])

fig.legend()
fig.footer("© European Centre for Medium-Range Weather Forecasts (ECMWF) \
           Source: www.ecmwf.int Licence: CC-BY-4.0 and ECMWF Terms of Use \
           (https://apps.ecmwf.int/datasets/licences/general/)", logo='ecmwf')

fig.show()
```

Note that plot produced using this dataset will slightly differ from one from Open Charts. This is due to different resolution of the data.  
The data used here is on 0.4x0.4 resolution, while high resolution data is 0.1x0.1 grid.
