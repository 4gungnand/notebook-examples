```python
from Magics import macro as magics


# Example reference


# Setting the coordinates of the geographical area

europe = magics.mmap(
    subpage_upper_right_longitude=60.,
    subpage_map_projection='cylindrical',
    subpage_lower_left_longitude=-20.,
    subpage_lower_left_latitude=30.,
    subpage_upper_right_latitude=60.,
    )

# Coastlines setting

coast = magics.mcoast(map_grid='on', map_grid_colour='tan',
               map_coastline_colour='tan',
               )

# Import andthe  data

obs = magics.mobs(obs_input_file_name='synop.bufr')




title = \
   magics.mtext(text_lines=["<font size='1'>Observation Plotting [Synop] </font>"
          , '',
          ], text_justification='left', text_font_size=0.5,
          text_colour='charcoal')



# To the plot

magics.plot( europe, obs, title, coast, )


```
