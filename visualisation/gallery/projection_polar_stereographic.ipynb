**Magics** supports several different projections, and they are controled using **mmap** funciton.

List of all **mmap** parameters you can find [in Magics documentation](https://confluence.ecmwf.int/display/MAGP/Subpage+-+Projection "Projection parameters").

```python
import Magics.macro as magics

# Setting the projection
projection = magics.mmap(
    subpage_upper_right_longitude  = 68.0, 
    subpage_upper_right_latitude   = 50, 
    subpage_lower_left_longitude   = -15., 
    subpage_lower_left_latitude    = 27.,
    subpage_map_vertical_longitude = 0., 
    subpage_map_projection         = "polar_stereographic",
    page_id_line                   = "off")

# Defining the coastlines
coast = magics.mcoast(
    map_coastline_resolution        = "high",
    map_coastline_colour            = "charcoal",
    map_coastline_land_shade        = "on",    
    map_coastline_land_shade_colour = "RGB(200, 200, 200)",
    map_grid                        = "on",
    map_grid_line_style             = "dot",
    map_grid_colour                 = "tan"
)

#Adding some text 
title = magics.mtext(
    text_lines = ["<font size='1'>Europe on Polar Stereographic projection</font>"], 
    text_justification = "left",
    text_font_size = 0.8,
    text_colour = "charcoal")
magics.plot(projection, coast, title)
```
