# Displaying grid point values

In this single cell notebook we will see how to display grid values together with isolines. Instead of value we could also plot a marker, or both.  

### Installing Magics

```python
import Magics.macro as magics

#Setting the geographical area
projection = magics.mmap(
    subpage_map_library_area = "on",
    subpage_map_area_name    = "central_europe",
    subpage_clipping         = "on",
    page_id_line             = "off"
)

#Setting the coastline
coast = magics.mcoast(    
    map_grid_line_style             = "dash",
    map_grid_colour                 = "grey")

#Loading GRIB file
tp = magics.mgrib(grib_input_file_name = "../../data/total_precipitation.grib")

#Defining the contour
tp_cont = magics.mcont(
    contour_grid_value_plot          = "on",
    contour_grid_value_height        = 0.50,
    contour_grid_value_plot_type     = "both",
    contour_grid_value_format        = "(i4)",
    contour_grid_value_colour        = "evergreen",
    contour_grid_value_marker_height = 0.5,
    contour_grid_value_marker_colour = "charcoal",
    contour_grid_value_lon_frequency = 3,
    contour_grid_value_lat_frequency = 3,
    contour_highlight                = "off",
    contour_line_colour              = "grey",
    contour_line_thickness           = 2,
    contour_level_selection_type     = "level_list",
    contour_level_list               = [0.5, 2, 4, 10, 20, 30, 50, 100, 250],
    contour_label                    = "off",
    legend                           = "off"
)

#Setting the title
title = magics.mtext( 
    text_lines          = ["Ploting isolines and grid point values",
                           "<grib_info key='name'/> -  STEP <grib_info key='step'/>",
                           "START TIME: <font colour='evergreen' style='bold'><grib_info key='base-date' format='%d %B %Y %H'/></font>   VALID TIME: <font colour='evergreen' style='bold'><grib_info key='valid-date' format='%d %B %Y %H'/></font>"],
    text_justification  = 'left',
    text_font_size      = 0.7,
    text_mode           = "automatic",
    text_colour         = "charcoal") 

#Plotting
magics.plot(projection, tp, tp_cont, coast, title)
```
