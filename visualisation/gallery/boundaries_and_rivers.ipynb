### Important parameters

#### To control coastline and land and sea shading use parameters:
- map_coastline - 'on' by default
- map_coastline_resolution - 'automatic' by default
- map_coastline_* - many parameters to control colour, thicknes, style of the coastline as well as shading of land and sea

#### To control plotting of boundaries use parameters:
- map_boundaries_* and
- map_disputed_boundaries - for the disputed boundaries

#### To control the grid lines use parameters:
- map_grid - 'on' by default
- map_grid_* to control the style, thickness, colour and frequency of grid lines

```python
import Magics.macro as magics

projection = magics.mmap(
    subpage_map_library_area = "on",
    subpage_map_area_name    = 'south_east_europe',
    page_id_line             = "off"
)


coast = magics.mcoast(
    map_boundaries                  = "on",
    map_disputed_boundaries         = "off",
    map_boundaries_colour           = "orangish_red",
    map_boundaries_thickness        = 2,
    map_coastline_resolution        = "high",
    map_coastline_colour            = "tan",
    map_coastline_land_shade        = "on",    
    map_coastline_land_shade_colour = "cream",
    map_grid                        = "on",
    map_grid_line_style             = "dot",
    map_grid_colour                 = "tan",
    map_grid_latitude_increment     = 4.00,
    map_grid_longitude_increment    = 4.00,   
    map_grid_latitude_reference     = 0.00,
    map_rivers                      = "on"
)



magics.plot(projection, coast)
```
