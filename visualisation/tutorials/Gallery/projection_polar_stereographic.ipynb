**Magics** supports several different projections, and they are controled using **mmap** funciton.

List of all **mmap** parameters you can find [in Magics documentation](https://confluence.ecmwf.int/display/MAGP/Subpage+-+Projection "Projection parameters").

```python
import Magics.macro as magics

# Setting the projection
projection = magics.mmap(
    subpage_lower_left_latitude   = 21.51,
    subpage_lower_left_longitude  = -37.27,
    subpage_upper_right_latitude  = 51.28, 
    subpage_upper_right_longitude = 65., 
    subpage_map_projection        = "polar_stereographic")

# Defining the coastlines
coast = magics.mcoast(
    map_coastline_resolution        = "high",
    map_coastline_colour            = "tan",
    map_coastline_land_shade        = "on",    
    map_coastline_land_shade_colour = "cream",
    map_grid                        = "on",
    map_grid_line_style             = "dot",
    map_grid_colour                 = "tan"
)

magics.plot(projection, coast)
```
