**Magics** supports several different projections.  
Some of them are:   
cylindrical, polar_stereographic, polar_north, mercator, mollweide, lambert, bonne etc.  
There are couple of satellite view and meteorology specific projections suported too, such as:  
meteosat, meteosat_57E, goes, goes_east etc. 

In this Notebook we will explore couple of them, and different ways of setting parameters.


### Import Magics

```python
import Magics.macro as magics
```

For start we will define some none projection parameters, which we will not change later.

```python
coast = magics.mcoast( 
            map_coastline_colour = "grey",
            map_coastline_land_shade =  "on", 
            map_coastline_land_shade_colour =  "cream", 
            map_grid =  "on", 
            map_grid_colour = "grey", 
            map_coastline_sea_shade  ="on", 
            map_coastline_sea_shade_colour=  "white", 
            map_label  ="off")
```

If you are a beginner, and want to focus on other things, Magics has several predefined projections.  
Let's check what they are.

```python
areas = magics.predefined_areas()
areas
```

```python
projection = magics.mmap(
    subpage_map_library_area = "on",
    subpage_map_area_name = u'south_west_europe'
)

magics.plot(projection, coast)
```

### Global cylindrical projection

Now lets see how we can use some of Magics projecitons.  
Global cylindrical projection is also Magics' default projection.

```python
projection = magics.mmap(
                  subpage_map_projection        = "cylindrical",
                  subpage_upper_right_longitude = 180.,
                  subpage_upper_right_latitude  = 90.,
                  subpage_lower_left_longitude  =  -180.,
                  subpage_lower_left_latitude   = -90.)

magics.plot(projection, coast)
```

### Cylindrical projection - Australia

```python
projection = magics.mmap(
                  subpage_upper_right_longitude = 190.00,
                  subpage_upper_right_latitude = -5.00,
                  subpage_lower_left_latitude = -55.00,
                  subpage_lower_left_longitude = 80.00,
                  subpage_map_projection = "cylindrical")

magics.plot(projection, coast)
```

### Polar stereographic - Europe

```python
projection = magics.mmap(
                  subpage_lower_left_latitude = 21.51,
                  subpage_lower_left_longitude =  -37.27,
                  subpage_upper_right_latitude =  51.28, 
                  subpage_upper_right_longitude =  65., 
                  subpage_map_projection =  "polar_stereographic")

magics.plot(projection, coast)
```

### Polar stereographic - North America

```python
projection = magics.mmap(
                  subpage_lower_left_latitude   = -5.00,
                  subpage_lower_left_longitude  = -140.00,
                  subpage_upper_right_latitude  = 30.00, 
                  subpage_upper_right_longitude = -15.00, 
                  subpage_map_vertical_longitude = -100.00,
                  subpage_map_projection =  "polar_stereographic"
)

magics.plot(projection, coast)
```

### Lambert projection

```python
projection = magics.mmap(
                  subpage_map_projection = "lambert",
                  subpage_lower_left_latitude = -0.00,
                  subpage_lower_left_longitude = -80.00,  
                  subpage_upper_right_latitude = 70.00,
                  subpage_upper_right_longitude = 160.00,
                  subpage_map_area_definition = "corners"
)

magics.plot(projection, coast)
```

### Mollweide projection

```python
projection = magics.mmap(subpage_map_projection = "mollweide",
                          subpage_map_area_definition = "full")

magics.plot(projection, coast)
```

### Tilted perspective (experimental)

```python
projection = magics.mmap(subpage_map_projection = 'tpers', 
                         subpage_frame = "off",
                         page_frame = "off",
                         subpage_map_projection_tilt = 0.,
                         subpage_map_projection_azimuth = 60.,
                         subpage_map_projection_view_latitude= 40.,
                         subpage_map_projection_view_longitude= 10.
                        )

magics.plot(projection, magics.mcoast())
```
