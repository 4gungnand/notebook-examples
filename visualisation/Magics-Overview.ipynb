# Magics Overview 


## What is Magics? 

Magics is an Open Source library written in C++, with an easy to use Python interface that helps you create your meteorological visualisations.  
The source code is available on [GitHub](https://github.com/ecmwf/magics).  
We have also created some notebooks examples to illustrate its  functionalities, also avalaible on [GitHub](https://github.com/ecmwf/notebook-examples) with a [Binder Option](https://mybinder.org/v2/gh/ecmwf/notebook-examples/master)


### Magics is meteorologically oriented 
It can visualise the main formats used by the community, giving easy visualisation options, and 
  * GRIB
  * NetCDF
  * BUFR 
  * from last year, we also added xarray.

### Magics provides a series of visual *Actions*
  * Contouring : Isolines/Shading
  * Wind Plotting : Arrows/Flags
  * Symbol plotting: Basic/WMO Weather and clouds symbols
  * Graph : ENSGrams
  * Coastlines plotting
  * Projection settings
  * Legend
  * Text

### And a large set of parameters for each of these *Actions* 
  * Isolines configuration
  * Colour 
  * Line Attributes
  * Font attributes 
  
## But let's so a simple Hello World example 

When creating a plot with Magics, you always have to start by defining the geographical area and the projection you want to use.

**Let's say we want to display temperature over Europe**

First we define the projection using the *mmap* Object.

```python
from Magics import macro as magics
```

```python
europe = magics.mmap(subpage_map_library_area = True, 
                     subpage_map_area_name = "europe",)

magics.plot(europe, magics.mcoast())
```

In this very first example, we have decided to use one of the predefined areas of Magics.  
These areas are defined in one the Magics configuration files,and can be extended. 
You can see the list of predefined area with : 

```python
for area in magics.predefined_areas():
    print (area)
```

But we can always define an own area in one the on the projection avalaible.

```python
area = magics.mmap(subpage_map_projection        = "cylindrical",
                  subpage_upper_right_longitude = 135.,
                  subpage_upper_right_latitude  = 50.,
                  subpage_lower_left_longitude  =  100.,
                  subpage_lower_left_latitude   = 20.,
                  page_id_line                  = "off" )
magics.plot(area, magics.mcoast())
```
