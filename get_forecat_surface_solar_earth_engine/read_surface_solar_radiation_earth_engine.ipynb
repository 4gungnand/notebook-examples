```python
import pandas as pd
from datetime import datetime

# Load the CSV file
file_path = '/home/prema/notebook-examples/ecmwf_solar_radiation_tokyo_hourly.csv'
data = pd.read_csv(file_path)

# Convert forecast_time from milliseconds to a readable datetime format
data['forecast_time'] = pd.to_datetime(data['forecast_time'], unit='ms')

# Display the processed data
data
```

```python
# Split the date and the hour
data['date'] = data['forecast_time'].dt.date
data['hour'] = data['forecast_time'].dt.hour
# Drop the original forecast_time column
data = data.drop(columns=['forecast_time'])

# Display the updated DataFrame
data
```
